Exploratory Data Analysis


```python
# whats the total revenue
Total_revenue=df["Sales"].sum()
Total_revenue
```

Total revenue is 1228455.8642

```python
# whats the most popular product
# value count will help to count the number of times a particular product was sold
Sold_product=df["Product Name"].value_counts().head(5)
Sold_product.sort_values(ascending=False)
```

The most popular product is nokia smart phone with caller id,motorola smart phone cordless,hoover stove red,motorola smart phone fullsize.
Recommendation: This products should always be available sincethis is what customers are buying the most.

```python
 # using the variable name
Sold_product.plot(kind="bar")
plt.title("Most popular product")
plt.xlabel("Product")
plt.ylabel("Frequency")
plt.show()
```

```python
# whats the average sale per product
Average_Sale=df["Sales"].mean()
Average_Sale
```

Average sale is 26137.358812765957

```python
#Whats the total quantity sold
Total_quantity=df["Quantity"].sum()
Total_quantity
```

Total quantity is 325

```python
# what is the most common customer segment
Customer_Segment=df["Segment"].value_counts()
Customer_Segment.sort_values(ascending=False)
```

The most common segment is consumer segment with 24 customer.
Recommendation: Consumer segment is seen as the most common segment.Tere should be given discountand rewards to make buy more.For other segments they should be reached out to to find out their preferred products.

```python
Customer_Segment.plot(kind="bar")
plt.title("The most common segment")
plt.xlabel("Segment")
plt.ylabel("Frequency")
plt.show()
```

```python
# The most profitable product category
Sales_by_category =df.groupby("Category")["Sales"].sum()
Sales_by_category.sort_values(ascending=False)
```

The most profitable category is technology with the highest amount of sales 491854.6952
Recommendation: Technology products should be available because it has the highest selling point.

```python
Sales_by_category.plot(kind="pie",autopct="%1.1f%%")
plt.legend()
plt.title("The most profitable category")
plt.show()
```

```python
# Whats the region that generates the highest sales
Sales_by_region=df.groupby("Region")["Sales"].sum()
Sales_by_region.sort_values(ascending=False)
```

The region that generates highest sales is south with amount of sales 235019.2022
Recommendation: It is seen that South region and Central region generates more sales, so this regions should have more of the top selling products.

```python
Sales_by_region.plot(kind="bar")
plt.title("The Region that generates more sales")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.show()
```

```python
# Top five best selling state
Sales_by_state=df.groupby("State")["Sales"].sum()
Sales_by_state.sort_values(ascending=False).head(5)
```

The best selling state is Tuscanywith amount of sales 111420.120 .
Recommendation:It is seen that Tuscany and England generates more sales, so this regions should have more of the top selling products.

```python
# Top five best selling product
Sales_by_product=df.groupby("Product Name")["Sales"].sum()
Sales_by_product.sort_values(ascending=False).head(5)
```

hoover stove, red is the highest selling product with amount of sales of 129838.548
Recommendation: The top 5 highest product should be focused more since it has the highest sales.

```python

df["Month"]=df["Order Date"].dt.month_name() # how to extract month
df.head()
```

```python
# whats the monthly sales trend
Sales_by_month=df.groupby("Month")["Sales"].sum()
Sales_by_month.sort_values(ascending=False)
```

There is an increase in August with 165564.2772 which also leads to the increase in September with  176165.8280
Recommendation: use the region to understands what happens in August and September and also what products sell more in this more in this month.

```python
Sales_by_month.plot(kind="line")
plt.title("The monthly sales trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```
